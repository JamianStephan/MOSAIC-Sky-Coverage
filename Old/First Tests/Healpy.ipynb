```python
import matplotlib.pyplot as plt
import numpy as np
import healpy as hp
```

```python
NSIDE = 2**13
print(
    "Approximate resolution at NSIDE {} is {:.2} deg".format(
        NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60
    )
)
```

```python
NPIX = hp.nside2npix(NSIDE)
print(NPIX)

```

```python
m = np.arange(NPIX)
#hp.mollview(m, title="Mollview image RING")
#hp.graticule()

```

```python
b = np.pi*0
a = np.pi*2
vec = hp.ang2vec(b, a)
```

```python
ipix_disc = hp.query_disc(nside=NSIDE, vec=vec, radius=np.radians(10/60))
```

```python
print(len(ipix_disc))
```

```python
m[ipix_disc]=m.max()
```

```python
m = np.arange(NPIX)
m[ipix_disc] = m.max()
hp.mollview(m, title="Mollview image RING")
```

```python
hp.gnomview(m, rot=[np.degrees(a), np.degrees(-b + np.pi/2)], title="GnomView",reso=0.05, xsize=1000)
```
