```python
indexes=[3,6,8,2,9,4]
pos=[1,1,3,3,6,6]
```

```python
chosen=[]
chosen.append(indexes[0])
```

```python
for i in range(1,len(indexes)):
    if abs(pos[0]-pos[i])>1:
        chosen.append(indexes[i])
        break

for o in range(i,len(indexes)):
    if abs(pos[0]-pos[o])>1 and abs(pos[i]-pos[o])>1:
        chosen.append(indexes[o])
        break
```

```python
print(chosen)
```
